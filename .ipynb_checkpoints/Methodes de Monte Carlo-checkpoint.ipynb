# Méthode d'acceptation-rejet

### Question 1:

On veut simuler la variable aléatoire réelle $X$ de densité:
$$
f(x) \propto exp\left(\frac{-\left(x - 1 \right)^2}{2} \right)|sin\left(\pi x \right)| \le exp\left(\frac{-\left(x - 1 \right)^2}{2} \right)
$$
Soit $g\left(x \right) = exp\left(\frac{-\left(x - 1 \right)^2}{2} \right) \forall x \in \mathbb{R}$, $g$ est la densité de la distribution normale $\mathcal{N}\left(1, 1 \right)$ qu'on sait simuler.

On peut donc utiliser la méthode d'acceptation-rejet pour simuler la variable aléatoire $X$

```python
def simulate(n):
    """
    Description
    --------------
    This function simulates n samples of the random variable X
    
    Parameters
    --------------
    n : Int, the number of samples.
    
    Returns
    --------------
    List of length n containing the n generated samples.
    """
    
    samples = []
    i = 0
    while i < n:
        y = random.normalvariate(1, 1)
        u = random.random()
        if u <= abs(math.sin(math.pi*y)):
            samples.append(y)
            i += 1
            
    return samples
    
```

Dans ce qui suit, on simule $50000$ réalisations de la variable aléatore $X$, on visualise l'histogramme de la distribution de notre échantillon ainsi que la densité ajustée pour comparer à la vrai densité $f$

```python
l = simulate(50000)
```

```python
import scipy.integrate as integrate

fig = plt.figure(figsize = (20, 5))

# Plot histogram and fitted probabilty distribution
plt.subplot(121)
sns.distplot(l, kde_kws = {'color' : 'r', 'label' : 'fitted distribution'})
plt.title('Histogram of the simulated variable')

# Plot the real probability distribution
pdf = lambda x : np.exp(-(x - 1)**2/2)*abs(np.sin(np.pi*x))
x_range = np.linspace(-4, 6, 1000)
plt.subplot(122)
plt.plot(x_range, pdf(x_range)/integrate.quad(pdf, -10, 10)[0], label = 'real distribution')

plt.legend()
plt.show()
```

La densité estimée à l'aide de l'histogramme de l'échantillon semble très proche de la vraie distribution.

Testons le temps qu'il faut pour avoir un grand échantillon ($10^6$ par exemple) de réalisations de $X$

```python
%%timeit
simulate(1e6)
```

Générer $10^6$ échantillons prend en moyenne moins de $3s$, ceci est dû au fait que $f$ et $g$ étant très similaires, la constante de domination est petit ($1$ ici) et on se retrouve avec un grand taux d'acceptation.

# Algorithme de Metropolis-Hastings

## Simulation sur $\mathbb{R}$

Dans cette partie on veut simuler des réalisations de la distribution $\chi^2(3)$ avec l'algorithme Metropolis-Hastings, pour cela on va utiliser la densité instrumentale suivante:
$$
q(y|x) = \frac{1}{s}\phi \left(\frac{y - x}{s} \right)
$$
avec $\phi$ la densité de la loi normale $\mathcal{N}\left(0, 1\right)$

### Question 1

```python
x0 = .0
nsamples = 50000
logpdf = lambda x : chi2.logpdf(x, 3) # Densité de la loi du chi2 à trois degrés de liberté.
proprnd = lambda s, x : x + s*np.random.normal() # Simulation du prochain point de la chaîne de Markov.
```

```python
def mhsim(s, x0, nsamples, logpdf, proprnd):
    
    """
    Description
    --------------
    This function simulates an MCMC with Metropolis-Hastings in order to get samples from the distribution
    of log-probability distribution function logpdf.
    
    Paramaters
    --------------
    s        : Float, the standard deviation of the proposal distribution.
    x0       : Float, the initial state.
    nsamples : Int, the number of samples to generate.
    logpdf   : function, the log-probability distribution of the target distribution.
    proprnd  : function, the function that proposes the next value of the markov chain.
    
    Returns
    --------------
    samples  : List containing the simulated samples.
    rho_mean : Float, the mean acceptance ratio.
    """
    
    x = x0
    samples = [x]
    accepted = 0    # Count the number of times we accepted a proposed point
    for i in range(1, nsamples):
        y = proprnd(s, x)
        rho = min(1, math.exp(logpdf(y) - logpdf(x)))
        if rho >= 1:
            x = y
            accepted += 1
            
        else:
            u = np.random.binomial(1, rho)
            x = u*y + (1 - u)*x
            accepted += 1*u
            
        samples.append(x)
        
    return samples, accepted/(nsamples - 1)
```

```python
samples1, rho_mean1 = mhsim(1e-3, x0, nsamples, logpdf, proprnd)
samples2, rho_mean2 = mhsim(1, x0, nsamples, logpdf, proprnd)
```

### Question 2

```python
print('acceptance rate for s = 1e-3 : %.3f' %rho_mean1)
print('acceptance rate for s = 1 : %.3f' %rho_mean2)
```

Pour $s = 10^{-3}$, il est normal que le taux d'acceptation soit si grand car on reste très proche du point courant de la chaîne de Markov, ceci implique que $\rho$ (le taux d'acceptation) reste proche de $1$

```python
fig = plt.figure(figsize = (20, 5))
plt.subplot(121)
plt.plot(range(nsamples), samples1)
plt.title('s = 1e-3', fontsize = 20)
plt.subplot(122)
plt.plot(range(nsamples), samples2)
plt.title('s = 1', fontsize = 20)
plt.show()
```

On remarque que pour $s = 1$ la chaîne de Markov "mélange" bien, ce qui n'est pas le cas pour $s = 10^{-3}$

Maintenant regardons les histogrammes des chaînes de Markov générées par le deux variances $s = 10^{-1}$ et $s = 1$ et comparons les densistés ajustées aux histogrammes à a vraie distribution de la loi $\chi^2(3)$

```python
x_axis = np.linspace(-1, 25, 100)

fig = plt.figure(figsize = (20, 5))

plt.subplot(121)
plt.plot(x_axis, chi2.pdf(x_axis, 3), label = 'target pdf')
sns.distplot(samples1, label = 'simulated distribution')
plt.legend()
plt.title('s = 1e-3', fontsize = 20)

plt.subplot(122)
plt.plot(x_axis, chi2.pdf(x_axis, 3), label = 'target pdf')
sns.distplot(samples2, label = 'simulated distribution')
plt.title('s = 1', fontsize = 20)
plt.legend()
plt.show()
```

Les histogrammes ci-dessus montrent que l'on approche bien la distribution $\chi^{2}(3)$ pour $s = 1$, ce qui n'est pas le cas pour $s = 10^{-3}$ . Ceci est normal car plus la variance est petite, plus on simule des points proches entre eux est plus le ratio d'acceptation est proche de $1$, ainsi comme le montre l'histogramme de la chaîne simulée avec $s = 10^{-3}$ on n'explore qu'une petite partie de l'espace au bout d'un temps long.

### Question 3

On veut maintenant règler automatiquement le taux d'acceptation, on a remarqué dans les expériences précédentes que $s$ l'écart-type de la distribution instrumentale utilisée influence beaucoup le taux d'acceptation. L'idée maintenant est de commencer avec un grand écart-type et de le diminuer au cours du temps jusqu'à atteindre un taux d'acceptation proche de $0.5$ (qui est un bon taux d'acceptation en dimension $1$) et ainsi on aura bien exploré l'espace pour trouver es régions de concentration de la densité cible. Dans l'implémentation, à chaque valeur de $s$ on calcule le taux d'acceptation sur une plage de 100 avant de faire diminuer $s$.

On adopte une stratégie exponentielle avec température pour diminuer $s$:
$$
s = s_f + (s_i - s_f)e^{-\frac{t}{\tau}}
$$
où $s_i$ et $s_f$ sont respectivement l'écart-type initial et l'écart-type final.

```python
s = 10
```

```python
def adapt_mhsim(x0, nsamples, logpdf, proprnd, s_init, s_final, decay_rate = 1e-3, window = 100, plot = False):
    
    """
    Description
    --------------
    This function simulates an MCMC with an adaptive Metropolis-Hastings in order to get 
    samples from the distribution of log-probability distribution function logpdf.
    
    Paramaters
    --------------
    x0         : Float, the initial state.
    nsamples   : Int, the number of samples to generate.
    logpdf     : function, the log-probability distribution of the target distribution.
    proprnd    : function, the function that proposes the next value of the markov chain.
    s_init     : Float, the initial standard deviation of the proposal distribution.
    s_final    : Float, the final standard deviation of the proposal distribution.
    decay_rate : Float, the decay rate of the standard deviation.
    window     : Int, length of the range on which we calculate the mean of the acceptance rate.
    
    Returns
    --------------
    samples  : List containing the simulated samples.
    rho_mean : Float, the mean acceptance ratio on a 100 range.
    s        : Float, The final standard deviation of the proposal distribution.
    """
    
    x = x0
    samples = [x]
    rho_values = []
    s = s_init
    updates = 0
    for i in range(1, nsamples):
        y = proprnd(s, x)
        rho = min(1, math.exp(logpdf(y) - logpdf(x)))
        if rho >= 1:
            x = y
            
        else:
            u = np.random.binomial(1, rho)
            x = u*y + (1 - u)*x
            
        samples.append(x)
        rho_values.append(rho)
        if len(rho_values) == window:
            rho_mean = np.mean(rho_values)
            if rho_mean < 0.5:
                updates += 1
                s = s_final + (s_init - s_final)*math.exp(-decay_rate*updates)
                rho_values = []
        
    return samples, rho_mean, s
```

```python
samples, rho_mean, s = adapt_mhsim(x0, nsamples, logpdf, proprnd, 10, 0.1, 1e-1)
```

```python
print('acceptance rate with annealing scheme : %.3f' %rho_mean)
print('final standard deviation : %.3f' %s)
```

```python
plt.plot(range(nsamples), samples)
plt.show()
```

```python
x_axis = np.linspace(-1, 25, 100)
plt.plot(x_axis, chi2.pdf(x_axis, 3), label = 'target pdf')
sns.distplot(samples, label = 'simulated distribution')
plt.show()
```

On doit faire attention à l'écart-type initial, s'il est très grand on risque de se placer dans un point de la chaîne de Markov où la densité la distribution cible est plate, il serait alors difficile de sortir d'une telle zone.

### Question 4

Dans cette question on veut diagnostiquer automatiquement un éventuel problème de convergence de la chaîne de Markov générée. L'idée est de comparer plusieurs chaînes générées à partir de différents points et voir si celles-ci engendrent des échantillons distribués selon la même distribution.

On se donne un mélange de deux lois normales $\mathcal{N}\left(0, 1 \right)$ et $\mathcal{N}\left(15, 0.5 \right)$ car c'est une distribution bimodale et représente donc la difficulté de rester coincer au voisinage de l'un des optimums pour la chaîne, on verra ainsi si deux chaînes générées à partir de deux points initiaux différents convergeront vers la même distribution invariante.

```python
unnorm_pdf = lambda x : norm.pdf(x, 0, 1) + norm.pdf(x, 15, 0.5)
M = integrate.quad(unnorm_pdf, -10, 20)[0]  # Normalizing constant
pdf = lambda x : (unnorm_pdf(x))/M
logpdf = lambda x : np.log(pdf(x))
proprnd = lambda s, x : x + s*np.random.normal()

x_range = np.linspace(-5, 20, 1000)
plt.plot(x_range, pdf(x_range))
plt.title('N(0, 1) + N(15, 0.5)')
plt.show()
```

```python
nsamples = 50000
samples_1, rho_mean_1, s_1 = adapt_mhsim(0., nsamples, logpdf, proprnd, 1, 0.1, 1e-1)
samples_2, rho_mean_2, s_2 = adapt_mhsim(20, nsamples, logpdf, proprnd, 1, 0.1, 1e-1)
plt.plot(x_range, pdf(x_range), label = 'target pdf')
sns.distplot(samples_1, label = 'simulated distribution 1')
sns.distplot(samples_2, label = 'simulated distribution 2')
plt.legend()
plt.show()
```

Il est évident que les distributions invariantes des deux échatillons générés par les deux chaînes de Markov sont différentes dans ce cas.

On peut utiliser le test de Kolmogorov-Smirnov pour vérifier si les deux échantillons proviennent de la même loi.

```python
p_value = ks_2samp(samples_1, samples_2)[1]
print("p-value = %.5e" %p_value)
```

Effectivement la p-valeur est nulle et donc on rejette largement l'hypothèse que les deux échantillons proviennent de la même loi. Ainsi même en ayant des chaînes qui convergent, avec cette méthode on peut vérifier si nos chaînes issues de deux points différents ont la même loi invariante.

# Problème d'estimation

On a les équations d'évolution suivantes pour modéliser l'évolution des quantités de matière dans deux compartiments:

$$
\begin{cases}
\dot{q}_1 = -(x_1 + x_3)q_1 + x_2 q_2 \\
\dot{q}_2 = x_1 q_1 - x_2 q_2 \\
q(x, 0) = (1, 0)^T
\end{cases}
$$

L'objectif de cette parte consiste à estimer le vecteur de paramètres $x = (x_1, x_2, x_3)^T$ à partir d'observations bruitées $y_i \sim \mathcal{N}\left(q_2(x, t_i), s^2 \right) \hspace{2mm}\forall t_i \in \{1, \ldots, 15\}$ afin d'obtenir une estimation de la quantité de,matière maximale dans le deuxième compartiment.

### Question 1

On construit d'abord le simulateur la quantité de matière $q_2(x, t_i) \hspace{2mm}\forall t_i \in \{1, \ldots, 15\}$, ce qui veut dire qu'à partir d'un vecteur de paramètre $x = (x_1, x_2, x_3)^T$ on résout le système dynamique.

```python
from scipy.integrate import odeint

y0 = np.array([1, 0]) # quantité de matière initiale dans les compartiments.
x = np.array([0.6, 0.35, 0.15])
t = np.arange(1, 16)
def model(y, t, x):
    M = np.array([-x[0] - x[2], x[1], x[0], -x[1]]).reshape((2, 2))
    return np.dot(M, y)

odeint(model, y0, t, args = (x,))  # Simulation des quantités de matière.
```

Construisons un vecteur d'observations $y = (y_1, \ldots, y_{15})$ avec le vecteur de paramètres $x^* = (0.6, 0.35, 0.15)^T$

```python
simulate_qts = lambda x, t, s : odeint(model, y0, t, args = (x,))[:, 1] + np.random.normal(0, s, len(t))
```

```python
x = np.array([0.6, 0.35, 0.15])
s_noise = 0.01
y = simulate_qts(x, np.arange(1, 16), s_noise)
```

### Question 3

On se donne une loi à priori uniforme $X = \left(X_1, X_2, X_2 \right) \sim \mathcal{U}\left(\left[0, 1 \right]^3 \right)$ . Calculons la loi à posteriori de $X$ étant données les observations $Y_1, \ldots, Y_{15}$ . 

On a $p\left(X|Y_1, \ldots, Y_{15} \right) \propto p(X)p\left(Y_1, \ldots, Y_{15}|X \right)$

on calcule la vraisemblance $p\left(Y_1, \ldots, Y_{15}|X \right) \propto \prod_{i=1}^{15}exp\left(-\frac{\left(y_i - q_2\left(x, t_i \right) \right)^2}{2s^2} \right)$ car $\forall i \in \{1, \ldots, 15 \} \hspace{2mm} y_i \sim \mathcal{N}\left(q_2\left(x, t_i \right), s^2 \right)$ avec $s>0$ un écart-type que nous choisissons pour le modèle.

Ainsi on a $p\left(X|Y_1, \ldots, Y_{15} \right) \propto \mathbb{I}_{[0, 1]}(x_1)\mathbb{I}_{[0, 1]}(x_2)\mathbb{I}_{[0, 1]}(x_3)\prod_{i=1}^{15}exp\left(-\frac{\left(y_i - q_2\left(x, t_i \right) \right)^2}{2s^2} \right)$

Dans la suite on travaillera avec la log-vraisemblance plutôt que la vraisemblance pour des raisons numériques.

```python
loglikelihood = lambda x, y, s : - ((y - odeint(model, y0, t, args = (x,))[:, 1])**2).sum()/(2*s**2)
```

```python
def posterior(x, y, s):
    
    for xi in x:
        if (xi < 0) or (xi > 1):
            return 0
        
    return np.exp(loglikelihood(x, y, s))
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

def plot_posterior(index, value, fig, s = 0.1, place = 131):
    
    ax = fig.add_subplot(place, projection='3d')
    # Make data.
    X1 = np.arange(0, 1, 0.01)
    X2 = np.arange(0, 1, 0.01)
    X1, X2 = np.meshgrid(X1, X2)
    Z = np.zeros(X1.shape)
    for i in range(X1.shape[0]):
        for j in range(X2.shape[1]):
            if index == 0:
                Z[i, j] = posterior([value, X1[i, j], X2[i, j]], y, s)
                
            elif index == 1:
                Z[i, j] = posterior([X1[i, j], value, X2[i, j]], y, s)
                
            elif index == 2:
                Z[i, j] = posterior([X1[i, j], X2[i, j], value], y, s)

    # Plot the surface.
    surf = ax.plot_surface(X1, X2, Z, cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)

    # Customize the z axis.
#     ax.set_zlim(-1.01, 1.01)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.e'))

    # Add a color bar which maps values to colors.
    cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
    cbar.ax.tick_params(labelsize = 20)
    ax.tick_params(axis='both', which='major', labelsize=20)
    if index == 0:
        ax.set_title('posterior distribution when setting x0 = %.2f' %value, fontsize=30)
        
    elif index == 1:
        ax.set_title('posterior distribution when setting x1 = %.2f' %value, fontsize = 30)
        
    elif index == 2:
        ax.set_title('posterior distribution when setting x2 = %.2f' %value, fontsize = 30)
        
```

Regardons ce que donne la loi à posteriori lorsque $s = 0.6$

```python
x = np.array([0.6, 0.35, 0.15])
s_noise = 0.5
y = simulate_qts(x, np.arange(1, 16), s_noise)
```

```python
fig = plt.figure(figsize = (50, 10))
plot_posterior(0, 0.6, fig, s = s_noise, place = 131)
plot_posterior(1, 0.35, fig, s = s_noise, place = 132)
plot_posterior(2, 0.15, fig, s = s_noise, place = 133)
plt.show()
```

```python
fig = plt.figure(figsize = (50, 10))
plot_posterior(0, 0.6, fig, s_noise, place = 131)
plot_posterior(1, 0.35, fig, s_noise, place = 132)
plot_posterior(2, 0.15, fig, s_noise, place = 133)
plt.show()
```

### Question 4

Ayant une loi à priori $X \sim \mathcal{U}\left([0, 1]^3 \right)$ suggère qu'on une utilise un random walk Metropolis-Hastings en prenant une loi instrumentale $Y|X \sim \mathcal{N}\left(X, \Sigma \right)$ où $\Sigma$ est une matrice de covariance diagonale isotropique.

```python
x0 = np.array([0.5, 0.25, 0.25])
nsamples = 10000
proprnd = lambda s, x : x + np.random.multivariate_normal(np.zeros(3), (s**2)*np.eye(3))
```

```python
def mh(x0, nsamples, proprnd, s_init, s_final, s_noise = 0.05, decay_rate = 1e-3):
    
    x = x0
    samples = [x]
    accepted = []
    rho_values = []
    s = s_init
    updates = 0
    for i in range(1, nsamples):
        x_next = proprnd(s, x)
        rho = min(1, posterior(x_next, y, s_noise)/(posterior(x, y, s_noise) + 1e-200))
        if rho >= 1:
            x = x_next
            accepted.append(x)
            
        else:
            u = np.random.binomial(1, rho)
            x = u*x_next + (1 - u)*x
            
        samples.append(x)
        rho_values.append(rho)
        if len(rho_values) == 100:
            rho_mean = np.mean(rho_values)
            if rho_mean < 0.7:
                updates += 1
                s = s_final + (s_init - s_final)*math.exp(-decay_rate*updates)
                rho_values = []
                        
    return samples, rho_mean, s, accepted
```

```python
samples, rho_mean, s, accepted = mh(x0, nsamples, proprnd, 0.2, 0.002, s_noise = 0.01, decay_rate = 1e-1)
```

```python
print('acceptance rate with annealing scheme : %.3f' %rho_mean)
print('final standard deviation : %.3f' %s)
```

On obtient une bonne estimation de la quantité de matière maximale dans le compartiment 2 à partir de notre estimation du vecteur de paramètre $x$ avec Metropolis-Hastings.
