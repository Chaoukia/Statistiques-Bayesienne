## II - Méthode d'acceptation-rejet

L'objectif est de pouvoir tirer des observations d'une variable aléatoire qui suit une loi de densité :$$f(x) \propto \exp \left(-\frac{1}{2}(x-1)^2\right)|\sin(\pi x)|$$ 

on normalise la densité $f$ pour avoir une égalité : $f(x) = \frac{1}{Z} h(x)$ avec $ Z = \int_{\mathbb{R}}\exp \left(-\frac{1}{2}(x-1)^2\right)|\sin(\pi x)|$ et $ h(x) = \exp \left(-\frac{1}{2}(x-1)^2\right)|\sin(\pi x)|$

on peut donc majorer notre densité, ce qui revient à majorer $h$ par la densité $g$ d'une gaussienne $\mathcal{N}(1,1)$ multipliée par une constante :

$$h(x) \le \sqrt{2\pi} \,g(x)$$ 


````python
import matplotlib.pyplot as plt

def acceptation_rejet(n):
    
    """
    méthode acceptation-rejet pour simuler une variable aléatoire de loi f
    
    renvoie : les réalisalisations, taux d'acceptance moyen.
    """
    
    samples = []
    
    h = lambda x :  np.exp(-0.5*(x-1)**2) * np.abs(np.sin(np.pi*x))
    g = lambda x : 1/np.sqrt(2*np.pi) * np.exp(-0.5*(x-1)**2)
    
    cte = np.sqrt(2*np.pi)
    
    counter, acceptance = 0, 0 
    while len(samples)<n:
        
        x = np.random.normal(1,1)
        u = np.random.uniform()
        
        ll_h = h(x)
        ll_g = g(x)
        
        counter += 1
        
        if u*cte*ll_g <= ll_h: 
            samples.append(x)
            acceptance += 1
    
    #on compare l'histogramme estimé par notre échantillon à la vraie densité
    
    taux = acceptance/counter
    x = np.linspace(-3,5,1000)
    
    plt.figure(figsize = (13,5))
    plt.subplot(1,2,1)
    plt.plot(x,h(x), label="la fonction h")
    plt.plot(x, cte*g(x), label="la borne supérieure cg")
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.hist(samples, bins=50, density = True, label="histogramme estimé")
    plt.suptitle("Estimation de la densité par acceptation-rejet : %d points tirés avec un taux d'acceptation %.2f."%(n, taux))
    plt.legend()
    plt.show()
    
    return samples, taux
````

````python
samples, taux = acceptation_rejet(n=1e5)
````

# III - Algorithme de Metropolis-Hastings

## Simulation sur $\mathbb{R}$ 

Nous voulons produire une réalisation d’une chaîne de Markov partant de $x_0=0$, de longueur $n_{samples}=5.10^4$ avec un noyau de transition de MH $p \sim \mathcal{X^2(3)}$ comme loi invariante. 

On construit la chaîne de Markov à l'aide de la densité instrumentale suivante :

$$q(y|x)=\frac{1}{s}\phi\left(\frac{y-x}{s}\right),\quad\textrm{où $\phi \sim \mathcal{N}(0,1)$}$$

Ce qui revient à dire que $q(y|x) \sim \mathcal{N}(x,s^2)$

La loi instrumentale étant symétrique, la probabilité d'acceptation de l'algorithme MH peut se résumer à : $\mathcal{P(x_k,y_k)} = min(1, \frac{p(y_k)}{p(x_k)} )$

**Question 1**

````python
from scipy import stats
import seaborn as sns

# densité de la distribution chi2 à 3 degrés de liberté
chi2 = lambda x : stats.chi2.pdf(x=x,df=3)
logchi2 = lambda x : stats.chi2.logpdf(x=x,df=3)

def mhsim(n=1e3, x0 = 0.2, s=1e-1, p = chi2, logp = logchi2):
    
    """
    Metropolis-Hastings pour simuler une variable aléatoire de loi chi2
    avec une gaussienne comme loi instrumentale.
    
    renvoie : les réalisalisations, taux d'acceptance moyen.
    """
    
    samples = [x0]
    ps_accept = []
    acceptances = []
    
    while len(samples) < n:
        
        acceptances.append(0)
        y = np.random.normal(x0,s)
        u = np.random.uniform()
        
        p_accept = min(1, np.exp(logp(y) - logp(x0)))
        ps_accept.append(p_accept)
        
        if u <= p_accept:
            x0 = y
            samples.append(x0) 
            acceptances[-1] += 1
            
    taux = np.mean(acceptances)
    plt.figure(figsize = (17,5))
    
    plt.subplot(1,2,1)
    plt.plot(samples, label="la chaîne de réalisations générées")
    plt.legend()
    
    x = np.linspace(0,20,1000)
    plt.subplot(1,2,2)
    
    sns.distplot(samples, label="estimation par histogramme")
    plt.plot(x, p(x), label="la vraie densité")
    
    plt.suptitle("Estimation de la densité par MH : écart-type: %.2f, points tirés: %d ,taux d'acceptation %.2f."%(s, n, taux))
    plt.legend()
    plt.show()
    
    return samples, taux
````

**Question 2**

````python
ss = [1e-1, 1, 3, 5]
n = 1e5
for s in ss :
    _ = mhsim(n = n, s = s)
````

__Discussion__

Nous remarquons que si $s$ est petit, une fois dans le support de la distribution, on ne fait qu'accepter les échantillons vu qu'on parcout doucement la distribution de probabilité vers les probabilité, et donc surement. on obtient donc un grand taux d'acceptation. L'algorithme ne mélange pas bien les points et donc on finit par avoir une convergence __lente__ et une mauvaise estimation.

en augmentant la valeur de $s$, on obtient une convergence plus rapide et des échantillons fidèles à notre distribution, puisque la densité instrumentale nous permet désormais de bien explorer le support de la distribution en se dirigeant plus rapidement vers les grandes valeurs de $p$, en gardant un taux d'acceptation élevé, puisque on reste très souvent inclus dans le support de notre distribution. On voit que l'algorithme __mélange bien les points__.

Finalement, quand l'écart-type devient très grand, on tombe sur les mêmes problèmes qu'avant, c'est à dire une convergence lente et une plus faible probabilité de tomber sur des points du support ce qui explique le taux d'acceptation faible.

**Question 3** : Réglage automatique du taux d'acceptation

au lieu de fixer l'écart-type de la distribution instrumentale, on peut adapter l'algorithme à ce qu'on obtient un taux d'acceptation **t_optimal** qu'on fixe au préalable après quelques itérations. Ce qu'on sait déjà, c'est que le taux d'acceptation est une fonction décroissante de l'écart-type. On créée un procédé iteratif ou pour chaque tranche de points tirés, on estime le taux d'acceptation moyen. Si ce taux est plus grand que notre **t_optimal**, on réduit l'écart-type et vice-versa.  On peut commencer par un grand écart-type (taux d'acceptation faible) puis le modifier au fur et à mesure dépendemment du taux estimé pour obtenir après quelques itérations le taux d'acceptation voulu.

La démarche est simple, pour la ième itération, si le taux est plus grand que le taux voulu, on soustrait $\frac{1}{i}$ de notre écart-type et vice versa. la série harmonique est semi-convergente, donc on peut trouver une susbstitution pour que cette série converge vers n'importe quel réel. (théorème de reimann)

````python
def mhsim_adaptatif(n=1e3, x0 = 0.2, s_init = 10, s_min = 1e-2, t_final = 0.5, tranche = 200, p = chi2, logp = logchi2):
    
    """
    variante adaptative du Metropolis-Hastings pour simuler une variable aléatoire de loi chi2
    avec une gaussienne comme loi instrumentale.
    
    renvoie : les réalisalisations, taux d'acceptation moyen.
    """
    
    samples = [x0]
    ps_accept = []
    acceptances = []
    std = s_init
    
    taux, i = [], 0
    
    while len(samples) < n:
        
        acceptances.append(0)
        y = np.random.normal(x0, std)
        u = np.random.uniform()
        
        p_accept = min(1, np.exp(logp(y) - logp(x0)))
        ps_accept.append(p_accept)
        
        if u <= p_accept:
            x0 = y
            samples.append(x0) 
            acceptances[-1] += 1
        
        
        if (len(acceptances) % tranche == 0) :
            taux.append(np.mean(acceptances[-tranche:]))
            i += 1
            if taux[-1] <= t_final : std -= s_init/(i+1)
            else : std += s_init/(i+1)
            
            std = max(std, s_min)
                
    plt.figure(figsize = (20,5))
    
    plt.subplot(1,3,1)
    plt.plot(samples, label="la chaîne de réalisations générées")
    plt.legend()
    
    x = np.linspace(0,20,1000)
    plt.subplot(1,3,2)
    
    sns.distplot(samples, label="estimation")
    plt.plot(x, p(x), label="cible")
    plt.legend(loc = 'best')
    
    plt.subplot(1,3,3)
    plt.plot(taux, label = "l'évolution du taux d'acceptation")
    plt.legend()
    
    plt.suptitle("Estimation de la densité par MH : s: %.3f, points tirés: %d ,taux d'acceptation final %.2f."%(std, n, np.mean(taux[-10:])))
    plt.show()
    
    return samples, taux
````

````python
_, taux = mhsim_adaptatif(n = 1e5, t_final = 0.82)
````

On obtient bien n'importe quel taux d'acceptation voulu avec cette méthode avec une précision de 0.01 pour 1e5 points tirés.

**Question 4** : Méthode de diagnostique

On peut voir que MH se comporte bien avec les densités de probabilités unimodales. Ceci dit, essayons de voir si MH permettrait de tirer d'une densité multimodale. La plus naive des distributions peut être un mélange de deux gaussiennes qu'on montre ci-dessous.

````python
x = np.linspace(0,15,1000)
s1, s2, mu1, mu2 = 1.0, 0.5, 3.0, 12.0 
gm = lambda x : stats.norm.pdf(x, loc=mu1, scale=s1) + stats.norm.pdf(x, loc=mu2, scale=s2)
loggm = lambda x : np.log(gm(x))
plt.plot(x, gm(x))
plt.title('mélange de deux gaussiennes : la distribution cible')
plt.show()
````

````python
xs = [0.0, 14.0]
n = 1e5
samples = []
for x in xs :
    output,_ = mhsim_adaptatif(n = n, x0 = x, p=gm, logp=loggm, s_init=5.0, t_final = 0.8)
    samples.append(output)
````

On remarque que dans cet exemple, on obtient des échantillons tirés des deux modes du mélange mais on a des chaines qui ne se ressemblent pas et donc on peut se demander si on eu convergence, c'est à dire que nos deux essaies nous ont donné la même distribution stationnaire finale. 

Pour le diagnostique, on se base sur la statistique de __Gelman-Rubin__ qui se base sur l'analyse des variances pour affirmer l'existence d'un problème de convergence. On calcule les deux variances intra et inter chaines et les compare pour voir si on obtient les memes statistiques pour dire que nos deux chaines convergent.

les étapes du diagnostique avec deux chaines simulées de longueur $2m$: 

- On supprime les n premiers points de nos deux chaînes.
- On calcule la variance inter et intra chaines : 
 - variance intra chaine : $W= \frac{std_1 + std_2}{2}$
 - on calcule la moyenne des deux chaines combinées $\mu$
 - variance inter chaine : $B= m((\mu_1 - \mu)^2 + (\mu_2 - \mu)^2)$

On cacule la variance finale comme une moyenne pondérée des deux quantités $W$ et $B$
$$Var= (1 - \frac{1}{m})W+ \frac{1}{m}B$$

la dernière statistique est donc $$R = \sqrt{\frac{Var}{W}} $$ et doit être très proche de 1 (à 0.1 près) pour confirmer la convergence de notre méthode. 

````python
m = int(n//2)
chain1, chain2 = samples

W = (np.std(chain1[m:])**2 + np.std(chain2[m:])**2)/2

mean1 = np.mean(chain1[m:])
mean2 = np.mean(chain2[m:])
mean = (mean1 + mean2)/2

B = m * ((mean1 - mean)**2 + (mean2 - mean)**2)
var_theta = (1 - 1/m) * W + (1/m)*B

print("Gelmen-Rubin Statistic: ", np.sqrt(var_theta/W))
````

On voit que dans cet exemple, on a pas une bonne convergence. Et on le voit aussi dans les graphes d'avant.
- On peut aussi comparer la distribution des échantillons de chaque chaine et de voir si les deux échantillons proviennent de la même distribution ( test de Kolmogorov-Smirnov )

## Un problème d'estimation

Nous sommes face à un système dynamique qui suit les équations d'évolution suivantes :

$$
\begin{cases}
\overset{.}{q_1} = -(x_1 + x_3)q_1 + x_2q_2\\
\overset{.}{q_2} = x_1q_1 - x_2q_2
\end{cases}
$$

Avec 
$$
q :
\begin{cases}
\mathbb{R}^3\times\mathbb{R}\rightarrow \mathbb{R}^2\\
(x, t) \rightarrow q(x,t)
\end{cases}
\textrm{avec} \quad q(x,0)=(1,0)^T \quad \forall x
$$



On crée la fonction  ```q2_gen``` : $(x,t)\rightarrow (q_2(x,t_1),q_2(x,t_2), ..., q_2(x,t_{15})$ dans la cellule au dessus.

````python
from scipy import integrate

def dynamics(q, t, x1, x2, x3):
    """
    fonction qui définie l'équadiff qui régit le système.
    """
    q1, q2 = q
    return [-(x1+x3)*q1 + x2*q2, x1*q1 - x2*q2]

def q2_gen(x, t_final=15):
    """
    genère q2(x,t) pour t allant de 1 à t_final
    """
    q0 = [1,0]
    t = np.arange(1,t_final+1)
    x1, x2, x3 = x
    q1, q2 = integrate.odeint(func=dynamics, y0=q0, t=t, args=(x1, x2, x3)).T
    return q2
````

Nos mesures sont bruitées :  $y_i \sim \mathcal{N}\left(q_2(x,t_i),s^2\right)$, $s>0$

Construisons un vecteur d’observations $y = (y_1, y_2, ..., y_{15})^T$ pour le paramètre $x^* = (0.6, 0.35, 0.15)^T$

Nous allons tout d'abord générer 15 points de $q_2$ pour le paramètre $x^*$, pour ensuite les bruiter pour obtenir nos mesures finales.

````python
def sample_normal(mu, std):
    """
    échantillon de la loi normal centrée en mu et d'ecart-type std
    """
    sample = np.vectorize(lambda mu, std : np.random.normal(loc=mu, scale=std))
    return sample(mu, std)

print("q2 simulé jusqu'à l'instant 15 :\n",q2,"\n")

# mesure bruitée
y = sample_normal(mu=q2, std=0.5)

print("Le vecteur d'observations y :\n",y,"\n")
````

on adopte la loi uniforme à priori  : $x\sim \mathcal{U}[0,1]^3$

$$p(x) = \overset{3}{\underset{i=1}{\Pi}}\mathbb{1}_{[0,1]}(x_i)$$

La loi à postériori est proportionnelle à la quantité suivante :

$$
\begin{align*}
p(x|y) &\propto p(y|x)p(x)\\
       &\propto \overset{15}{\underset{i=1}{\Pi}}\mathcal{N}\left(q_2(x,t_i),s^2\right) \overset{3}{\underset{j=1}{\Pi}}\mathbb{1}_{[0,1]}(x_j)
\end{align*}
$$

Pour éviter tout problème de statibilité numérique, on passe au domaine logarithmique pour faire nos calculs avant de revenir au domaine original avec une exponentilelle. La log loi à posteriori s'écrit comme suit :

$$
\begin{align*}
\log(p(x|y)) &\propto \log(p(y|x)) + \log(p(x))\\
             &\propto -\frac{1}{2s^2}\overset{15}{\underset{i=1}{\sum}}\left(y_i-q_2(x,t_i)\right)^2 + \overset{3}{\underset{j=1}{\sum}}\log(\mathbb{1}_{[0,1]}(x_j))
\end{align*}
$$



````python
from mpl_toolkits import mplot3d
from matplotlib import cm

std = 0.5

def proba_a_posteriori(x1,x2,x3,y):
    """
    Densité de la probabilité à postériori de x
    """
    q2 = q2_gen(x=[x1,x2,x3], t_final=len(y))
    
    lognormal = np.vectorize(lambda x,mu,std : -0.5*(x-mu)**2/std**2)
    
    return np.exp(np.sum(lognormal(y,q2,std=std)))*np.prod([xi >= 0 and xi <=1 for xi in [x1,x2,x3]])

def plot_3D(x01, x02, x_fixed=1):
    
    x_true = [0.6, 0.35, 0.15]
    
    if x_fixed==1:
        x1,x2,x3 = x_true[0],x01,x02
    if x_fixed==2:
        x1,x2,x3 = x01,x_true[1],x02
    if x_fixed==3:
        x1,x2,x3 = x01,x02,x_true[2]
            
    return proba_a_posteriori(x1,x2,x3,y)

x01 = np.linspace(-0.1, 1.1, 50)
x02 = np.linspace(-0.1, 1.1, 50)

X01, X02 = np.meshgrid(x01, x02)

f = plt.figure(figsize=(15,5))

Z = np.vectorize(plot_3D)(X01, X02, x_fixed=1)
ax1 = f.add_subplot(1, 3, 1, projection='3d')
ax1.plot_surface(X01, X02, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
ax1.set_title("Loi à postériori | x1 fixé à 0.6")

Z = np.vectorize(plot_3D)(X01, X02, x_fixed=2)
ax2 = f.add_subplot(1, 3, 2, projection='3d')
ax2.plot_surface(X01, X02, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
ax2.set_title("Loi à postériori | x2 fixé à 0.35")

Z = np.vectorize(plot_3D)(X01, X02, x_fixed=3)
ax3 = f.add_subplot(1, 3, 3, projection='3d')
ax3.plot_surface(X01, X02, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
ax3.set_title("Loi à postériori | x3 fixé à 0.15")

plt.tight_layout()
plt.show()
````

````python
import seaborn as sns

def MH(obs, x0, n=5e4, var=5, plot=True, decay=False):
    """
    Construit une réalisation d'une chaîne de Markov partant de x0
    avec la loi instrumentale q(y|x) ~ N(x,s^2) pour une loi cible X^2(3)
    """
    samples = [x0]
    taux = []
    var_start, var_stop, decay_step, T = var, 0.5, 0, 10
    
    # Densité de la distribution cible
    p = lambda x : proba_a_posteriori(x1=x[0], x2=x[1], x3=x[2], y=obs)
    
    while len(samples)<n:
        y = np.random.multivariate_normal(mean=x0, cov=np.eye(len(list(x0)))*var)
        
        # La distribution q est symétrique
        P_xy = 1 if p(y)>=p(x0) else p(y)/p(x0)
        taux.append(P_xy)

        if np.random.uniform() <= P_xy:
            samples.append(y)
            x0 = y
                                          
        if decay and len(taux)%100==0 and len(taux) > 1:
            mean_acceptation = np.mean(taux[-100:])
            if mean_acceptation < 0.2:
                decay_step += 1
                #print("s",end="\t")
                var = (var_start-var_stop)*np.exp(-decay_step/T) + var_stop
                #print("s")
        
    
    if plot:
        f, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(16,4))
        x_true = [0.6, 0.35, 0.15]
        
        if not decay:
            plt.suptitle("Taux d'acceptation pour s = {} : {:.2f}%".format(var,np.nanmean(taux)*100), y=1.1, fontsize=14)
        else:
            plt.suptitle("Taux d'acceptation pour s allant exponnetiellement de {} à {:.2f} : {:.2f}%".format(var_start,var,np.nanmean(taux)*100), y=1.1, fontsize=14)
        
        
        sns.distplot(np.asarray(samples).T[0], ax=ax1)
        ax1.axvline(x=x_true[0], color="red", linestyle="--", label="Vraie valeur x1")
        ax1.set_title("Chaîne de Markov générée pour x1")
        
        sns.distplot(np.asarray(samples).T[1], ax=ax2)
        ax2.axvline(x=x_true[1], color="red", linestyle="--", label="Vraie valeur x2")
        ax2.set_title("Chaîne de Markov générée pour x2")
        
        sns.distplot(np.asarray(samples).T[2], ax=ax3)
        ax3.axvline(x=x_true[2], color="red", linestyle="--", label="Vraie valeur x3")
        ax3.set_title("Chaîne de Markov générée pour x3")
        
        plt.tight_layout()

        plt.show()
    
    return np.asarray(samples), taux
````

````python
res, taux = MH(obs=y, x0=[0,0,0], var=2, plot=True, n=1000, decay=True)
````
